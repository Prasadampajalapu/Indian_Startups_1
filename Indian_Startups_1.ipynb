```python
import pandas as pd
import operator

df=pd.read_csv('startup_funding.csv',encoding='utf-8',skipinitialspace=True)
df
```

```python
def set_dictionary(array):
    d={}
    #d holds the investers name as key and value as count of that invester
    for i in array:
        if ',' not in i and i!='':
            d[i]=d.get(i,0)+1
        else:
            #split the investers name list
            string=i.strip().split(',')
            
            for j in string:
                if j!='':
                    d[j.strip()]=d.get(j.strip(),0)+1
    return d
            
```

```python
df.dropna(inplace=True,subset=['InvestorsName'])
dic=set_dictionary(df['InvestorsName'])
dic
```

```python
k=sorted(dic.items(),key=operator.itemgetter(1),reverse=True)[:1]
k
```

```python
import pandas as pd
import operator

df=pd.read_csv('startup_funding.csv',encoding='utf-8',skipinitialspace=True)

def set_dictionary(array):
    d={}
    #d holds the investers name as key and value as count of that invester
    for i in array:
        if ',' not in i and i!='':
            d[i]=d.get(i,0)+1
        else:
            #split the investers name list
            string=i.strip().split(',')
            
            for j in string:
                if j!='':
                    d[j.strip()]=d.get(j.strip(),0)+1
    return d

df.dropna(inplace=True,subset=['InvestorsName'])
dic=set_dictionary(df['InvestorsName'])

k=sorted(dic.items(),key=operator.itemgetter(1),reverse=True)[:1]
print(k[0][0],k[0][1])
```
